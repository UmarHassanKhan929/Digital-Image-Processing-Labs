#Task #1: Image Negatives

Implement a function for displaying negative of an input image. Note that the function must handle binary, grayscale, and RGB images

```python
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img = Image.open("lena_color.gif")
```

```python
def color_to_neg(img):
  img_data = np.array(img)
  img_data = 256 - 1 - img_data
  neg_img = Image.fromarray(img_data)
  return neg_img
```

```python
img.convert('RGB').save("lena_color.jpg")
img.convert('L').save("lena_gray.jpg")
```

```python
img = Image.open("lena_color.jpg")
```

```python
rgb_img = color_to_neg(img)
```

```python
imshow(rgb_img)
```

```python
img = Image.open("lena_gray.jpg")
```

```python
gray_img = color_to_neg(img)
```

```python
imshow(gray_img,cmap='gray')
```

#Task #2: Image Gradients

```python
import numpy as np
from PIL import Image,ImageFilter
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img = Image.open("lena_gray.jpg")
imshow(img,cmap='gray')
```

```python
img_data = np.array(img)
img_data.shape
```

```python
for x in range(0,img.size[0]-1):
  for y in range(0,img.size[1]-1):
    img.putpixel((x,y),abs(img.getpixel((x+1,y))-img.getpixel((x,y)))) 
```

```python
imshow(img,cmap='gray')
```

#Task #3: Bit Plane Slicing

Perform bit slicing of an 8 bit greyscale image as discussed in the lecture. Start from the least significant bit and move towards the most significant bit. 

```python
import numpy as np
from PIL import Image,ImageFilter
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img = Image.open("bitplane.tif")
img.convert('L').save("bitplane.jpg")
```

```python
imga = Image.open("bitplane.jpg")
imshow(imga,cmap='gray')
```

```python
img_data = np.array(imga)

zero_bit_img = (img_data>>0)&1
zero_img = Image.fromarray(zero_bit_img)

one_bit_img = (img_data>>1)&1
one_img = Image.fromarray(one_bit_img)

two_bit_img = (img_data>>2)&1
two_img = Image.fromarray(two_bit_img)

three_bit_img = (img_data>>3)&1
three_img = Image.fromarray(three_bit_img)

four_bit_img = (img_data>>4)&1
four_img = Image.fromarray(four_bit_img)

five_bit_img = (img_data>>5)&1
five_img = Image.fromarray(five_bit_img)

six_bit_img = (img_data>>6)&1
six_img = Image.fromarray(six_bit_img)

seven_bit_img = (img_data>>7)&1
seven_img = Image.fromarray(seven_bit_img)
```
