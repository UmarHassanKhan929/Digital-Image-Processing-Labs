```python
import numpy as np
import cv2 as cv
from PIL import Image,ImageOps
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = Image.open('30 (2).jpg')
img
```

```python
img_gray = img.convert('L')
img_gray.save('gray.png')
img_gray
```

```python
img_array = np.array(img_gray)
img_array.shape
```

```python
img_array = cv.copyMakeBorder(img_array, 2, 2, 2, 2,cv.BORDER_REPLICATE)
```

```python
output_arr = np.zeros((img.size[1],img.size[0]),dtype='uint8')
```

```python
for x in range(2,img_array.shape[0]-3):
  for y in range(2,img_array.shape[1]-3):

    max_gray = max(img_array[x-1][y-1],img_array[x][y-1],img_array[x+1][y-1],img_array[x-1][y],img_array[x][y],img_array[x+1][y],img_array[x-1][y+1],img_array[x][y+1],img_array[x+1][y+1])
    min_gray = min(img_array[x-1][y-1],img_array[x][y-1],img_array[x+1][y-1],img_array[x-1][y],img_array[x][y],img_array[x+1][y],img_array[x-1][y+1],img_array[x][y+1],img_array[x+1][y+1])
    
    inten_level = max_gray - min_gray
    
    output_arr[x-2][y-2] = inten_level
```

```python
new_img_array = output_arr.astype(np.uint8) 
new_img  = Image.fromarray(new_img_array)
new_img
```

```python
new_img.save("filtered.png")
```
