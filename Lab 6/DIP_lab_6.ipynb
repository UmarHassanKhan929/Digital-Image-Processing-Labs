#Histogram Equalization

a)	Write a program that shows the histogram for a given image.  For displaying the histogram, you may use Matplotlib

```python
import numpy as np
from PIL import Image,ImageOps
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = Image.open("image_lab6.tif")
```

```python
img
```

```python
img_array = np.array(img)
```

```python
flattened_data = img_array.ravel() 
plt.hist(flattened_data,256,[0,256])  
plt.show()  
```

b)	Write a program that equalizes the histogram of a given image. Consider the formula,

```python
hist,bins = np.histogram(flattened_data,256,[0,256])
cdf = hist.cumsum()
```

```python
p_hist=[0.0]*256
for i in range(hist.shape[0]):
  p_hist[i] = hist[i]/float(cdf.max())
```

```python
p_hist = np.array(p_hist)
p_hist=p_hist.cumsum()
```

```python
for i in range(255):
  p_hist[i] = 255*p_hist[i]
```

```python
img_new = p_hist[flattened_data]
img_new
```

```python
img_new = np.reshape(img_new, img.size)
```

 c) Show the histograms before and after equalization. Again consult Matplotlib. Opencv can also be used for some parts.

```python
new_img_array = np.array(img_new)
```

```python
new_flattened_data = new_img_array.ravel() 
plt.hist(new_flattened_data,256,[0,256])  
plt.show()  
```

```python
img_new = new_img_array.astype(np.uint8)
img_new = Image.fromarray(img_new)
```
