#Task #1: Image Binarization using a predefined global threshold


Take an RGB image (preferably from the provided ones) and convert it to binarized form (in 0/1 form) by defining a single global threshold. Repeat the experiment with the three provided images and identify why a single global binarization threshold may not be applicable in a wide variety of application scenarios.
Note: You have to perform task 1 on following images  B1,B2,B3


```python
#importing libraries
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img = Image.open("B1.png") #mounting image
img #viewing image
```

```python
img_array = np.array(img) #converting to array
```

```python
def rgb_to_gray(i_array): #function to convert the rgb to grayscale image
  gray_img_manual = np.zeros((i_array.shape[0],i_array.shape[1]))
  for i in range(i_array.shape[0]):
    for j in range(i_array.shape[1]):
      r,g,b = i_array[i][j]
      gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
      gray_img_manual[i][j]=gray
  return gray_img_manual
```

```python
gray_data = rgb_to_gray(img_array)
```

```python
def Binarizing(img_data,threshold): #function to convert the grayscale image to binary image
  for x in range(img_data.shape[0]-1):
    for y in range(img_data.shape[1]-1):
      if img_data[x][y] < threshold: # condition for threshold
        img_data[x][y] = 0
      else:
        img_data[x][y] = 255
  return img_data
```

```python
binary_data = Binarizing(gray_data,180) #getting the binary data
```

```python
binary_data = binary_data.astype(np.uint8) #changing data type to match image type
binary_img  = Image.fromarray(binary_data) #converting binary data to binary image
binary_img
```

#Task #2: Create Intensity Histogram from a Greyscale image

```python
#importing libraries
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img = Image.open("B1.png") #mounting image
img_array = np.array(img) #converting image to array
gray_data = rgb_to_gray(img_array) #converting rgb data to grayscale array
flattened_data = gray_data.ravel() #flattening the grayscale image array i.e. coverting into 1d array
plt.hist(flattened_data,256,[0,256]) #plotting graph, data given, max range given,  
plt.show() #displaying graph
```
