#Task #1: Effect of averaging and the size of averaging filters

```python
import numpy as np
import cv2 as cv
import math
from PIL import Image,ImageOps
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = Image.open("smoothing.tif")
img
```

```python
img_array = np.array(img)
```

```python
kernel = np.ones((15,15),dtype='uint8')
```

```python
soc_list = kernel.ravel()
soc = np.sum(soc_list)
```

```python
padding = kernel.shape[0] - 2
```

```python
img_array = cv.copyMakeBorder(img_array, padding, padding, padding, padding,cv.BORDER_REPLICATE)
```

```python
output_img = np.zeros((img.size[0]+padding,img.size[1]+padding),dtype='uint8')
```

```python
for x in range(0,img_array.shape[0]-1):
  for y in range(0,img_array.shape[1]-1):
    if (x <= padding or y<= padding):
      continue
    else:
      if (x>=img.size[0]-1 or y>=img.size[1]-1):
        break

      aoi = img_array[x-math.ceil((padding/2)):x+math.ceil((padding/2))+1, y-math.ceil(padding/2):y+math.ceil((padding/2))+1]
      k = ((aoi * kernel).sum())/soc
      output_img[x][y]=k
```

```python
output_img = output_img.astype(np.uint8)
output = Image.fromarray(output_img)
```

```python
final = np.array(img)
for x in range(padding,output_img.shape[0]-1):
  for y in range(padding,output_img.shape[1]-1):
    if (x>=img.size[0]-1 or y>=img.size[1]-1):
      break
    final[x-padding][y-padding]=output_img[x][y]
```

```python
output_img = final.astype(np.uint8)
output = Image.fromarray(output_img)

output
```
