```python
import numpy as np
import skimage.util 
import skimage
from skimage import filters
import cv2 as cv
from PIL import Image,ImageOps
from matplotlib.pyplot import imshow,title,figure
from matplotlib import pyplot as plt
import math
from scipy import ndimage, misc
import scipy
import statistics
%matplotlib inline
```

```python
img = Image.open('Picture2.jpg')
imshow(img)
```

```python
image = cv.imread('Picture2.jpg')
kernel = np.ones((2, 2), np.uint8)
changed = cv.dilate(image, kernel,iterations=2)
imshow(changed,cmap='gray')
```

```python
th,img_th = cv.threshold(changed, 50, 180, cv.THRESH_BINARY)
imshow(img_th,cmap='gray')
```

```python
gray = cv.cvtColor(img_th, cv.COLOR_BGR2GRAY)
imshow(gray,cmap='gray')
```

```python
kernel = np.ones((7, 7), np.uint8)
changed = cv.erode(gray, kernel)
imshow(changed,cmap='gray')
```

```python
blurr = cv.blur(changed,(3,3))
imshow(blurr,cmap='gray')
```
